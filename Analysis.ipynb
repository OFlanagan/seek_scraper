```python
def clean_string(string):
    pattern = re.compile(r'<(.*?)>')
    string = pattern.sub("",string)
    pattern = re.compile(r'\n{2,}')
    string = pattern.sub('\n',string)
    pattern = re.compile(r' {2,}')
    string = pattern.sub(' ',string)
    return string

def clean_non_standard_string(string):
    pattern = re.compile(r'\n{2,}')
    string = pattern.sub(' ',string)
    pattern = re.compile(r' {2,}')
    string = pattern.sub(' ',string)
    pattern = re.compile(r'\n')
    string = pattern.sub('NEWLINE',string)
    pattern = re.compile(r'\((.*?)\)')
    string = pattern.sub('',string)
    pattern = re.compile(r'\{(.*?)\}')
    string = pattern.sub('',string)
    pattern = re.compile(r'window.SEEK.*')
    string = pattern.sub('',string)
    pattern = re.compile(r'Report this job advertBe careful.*')
    string = pattern.sub("",string)
    pattern = re.compile(r'.*\}')
    string = pattern.sub("",string)
    pattern = re.compile(r'.*for saleVolunteeringAUNZ')
    string = pattern.sub("",string)
    pattern = re.compile(r'NEWLINE')
    string = pattern.sub('\n',string)
    pattern = re.compile(r'\.')
    string = pattern.sub(".\n",string)
    pattern = re.compile(r"\\")
    string = pattern.sub(" \ ",string)
    return string

def detect_non_standard_string(string):
    pattern = re.compile(r'\(function\(l,d,j,t\)')
    return pattern.search(string)!=None
    

df["tidied_main_text"] = [clean_non_standard_string(string)  if detect_non_standard_string(string) == True else clean_string(string) for string in df["main_text"]]
```

```python
for i, string in enumerate(df["tidied_main_text"]):
    print(i)
    print(string)
```

```python
pattern = re.compile(r".*year.*")
[pattern.search(string) for string in df["tidied_main_text"] if pattern.search(string) != None]
```

```python
i = 5
pattern = re.compile(r"[0-9](.*?)years(.*?)")

pattern.search(df["tidied_main_text"][i]).group(0)
```

```python
pattern = re.compile(r"\..*year.*")
[pattern.search(string).group(0) if pattern.search(string) != None else pattern.search(string) for string in df["tidied_main_text"]]
```

```python
pattern = re.compile(r".*year.*")
df["year_text"] = [pattern.search(string).group(0) if pattern.search(string) != None else pattern.search(string) for string in df["tidied_main_text"]]
```

```python
pattern = re.compile(r".*((c|C)itizen).*")
df["citizen"] = [pattern.search(string).group(0) if pattern.search(string) != None else pattern.search(string) for string in df["tidied_main_text"]]
```

```python
pattern = re.compile(r".*((p|P)ython).*")
df["python"] = [pattern.search(string).group(0) if pattern.search(string) != None else pattern.search(string) for string in df["tidied_main_text"]]
pattern = re.compile(r".*(SQL).*")
df["SQL"] = [pattern.search(string).group(0) if pattern.search(string) != None else pattern.search(string) for string in df["tidied_main_text"]]
pattern = re.compile(r".*(( |/|,){1,}R( |/|,){1,}).*")
df["R"] = [pattern.search(string).group(0) if pattern.search(string) != None else pattern.search(string) for string in df["tidied_main_text"]]
```
